```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
n = 20
p = 0.5
x = np.arange(0, n + 1)
```

```python
# Manual PMF function
def binomial_pmf_manual(k, n, p):
 return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
```

```python
# Calculate PMF manually
pmf = [binomial_pmf_manual(k, n, p) for k in x]
```

```python
# Plot
plt.figure(figsize=(10, 6))
bar_width = 0.8
plt.bar(x - bar_width/2, pmf, width=bar_width, color='blue', label='Manual Calculate')
plt.title(f'Binomial Distribution (Without Builtin Function) - n={n}, p={p}')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import binom
```

```python
# Calculate PMF using built-in function
pmf = binom.pmf(x, n, p)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x + bar_width/2, pmf, width=bar_width, color='red', alpha=0.7, label='Built')
plt.title(f'Binomial Distribution (Using Built-in Function(SciPy)) - n={n}, p={p}')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
