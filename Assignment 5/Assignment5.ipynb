Assignment No-05 :- Assignment based on plotting a normal distribution curve using python,

Without Using Inbuilt Functions

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
mean = 75
std_dev = 10
```

```python
# Generate x values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
```

```python
# Calculate y values manually using the PDF formula
y = [(1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(-((val - mean)**2) / (2 * std_dev**2)) for val in x]
```

```python
# Plot
plt.plot(x, y, color='red', linewidth=3)
plt.title('Normal Distribution Curve (Without Inbuilt Function)')
plt.xlabel('Data Values')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

Using Inbuilt Functions

```python
# Plotting Normal Distribution using inbuilt functions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate normally distributed data
mean = 75
std_dev = 10
data = np.random.normal(mean, std_dev, 1000)
```

```python
# Create range for x-axis
x = np.linspace(min(data), max(data), 100)
```

```python
# Calculate probability density function
pdf = norm.pdf(x, mean, std_dev)
```

```python
# Plot
plt.plot(x, pdf, color='blue', linewidth=3)
plt.title('Normal Distribution Curve (Inbuilt Functions)')
plt.xlabel('Data Values')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
